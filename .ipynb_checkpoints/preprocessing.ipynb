```python
#importing libraries
import pandas as pd
import json
```

Cleaning Match Summary

```python
with open('data/Raw/JSON Files/t20_wc_match_results.json') as f:
    data = json.load(f)
    
df_matches = pd.DataFrame(data[0]['matchSummary'])
df_matches.head()
```

```python
df_matches.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_matches.head()
```

```python
match_id_dict = {}

for index, row in df_matches.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_id_dict[key1] = row['match_id']
    match_id_dict[key2] = row['match_id']
```

```python
df_matches.to_csv('data/Processed/match_summary.csv', index = False)
```

Cleaning Batting Summary

```python
with open('data/Raw/JSON Files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    
    records = []
    
    for record in data:
        records.extend(record['battingSummary'])
    
df_batting = pd.DataFrame(records)
df_batting.head()
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: 'out' if len(x) > 0 else 'not out')
df_batting.drop(columns=['dismissal'], inplace=True)
df_batting.head()
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head()
```

```python
df_batting['match_id'] = df_batting['match'].map(match_id_dict)
df_batting.head()
```

```python
df_batting.to_csv('data/Processed/batting_summary.csv', index = False)
```

Cleaning Bowling Summary

```python
with open('data/Raw/JSON Files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    
    records = []
    
    for record in data:
        records.extend(record['bowlingSummary'])
    
df_bowling = pd.DataFrame(records)
df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_id_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('data/Processed/bowling_summary.csv', index = False)
```

Cleaning Players Information

```python
with open('data/Raw/JSON Files/t20_wc_player_info.json') as f:
    data = json.load(f)
    
df_players = pd.DataFrame(data)
df_players.head()
```

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head()
```

```python
df_players.to_csv('data/Processed/players.csv', index = False)
```
